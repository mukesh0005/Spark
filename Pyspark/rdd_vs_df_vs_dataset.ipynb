```python
from pyspark.sql import SparkSession;
from pyspark.context import SparkContext;
spark = SparkSession \
    .builder \
    .appName("rdd vs df") \
    .getOrCreate()

sc = spark.sparkContext
```

### Type safety
- rdds are not type safety at run time but compile time is fine
- dfs are not 

```python
rdd = sc.parallelize([1, 2, 3, "four"])
type(rdd)
```

```python
rdd.map(lambda x: x * 2).collect()
```

```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType
```

```python
data = [(1, "Alice"), ("two", "Bob")]
```

```python
## PySpark DataFrames enforce some type safety using schemas but cannot prevent incorrect types at compile-time.
schema =  StructType([
    StructField("id", IntegerType(), True),
    StructField("name", StringType(), True),
])

spark.createDataFrame(data,schema=schema)
```

```python
type_safety_df = spark.createDataFrame(data)
```

```python
type_safety_df.show()
# we get error becos it is compile time
```

### Performance
- rdds require more no of operations
- dfs have catalyst optimizer which is an advantage

### Memory management

- rdds use serialization
- no columnar storage
- no in memory compression
- no off- heap memeory

### Serialization

```python
## rdds 
## java or kyro (better)
data = [("Alice", 30), ("Bob", 25), ("Charlie", 35)]
```

```python
rdd = sc.parallelize(data)
import pickle

## java serialization
serialized_rdd = rdd.map(lambda x: pickle.dumps(x)).collect()
print(serialized_rdd)
```

### Schema Enforcement
